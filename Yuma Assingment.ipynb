# Data Visualization

```python
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

### Sales Over time

```python
# grouping by Transaction Date
sales_by_date=df.groupby('Transaction_Date')['TotalAmount'].sum().reset_index()

qty_by_date=df.groupby('Transaction_Date')['Quantity'].sum().reset_index()
```

```python
# Create figure and axis
fig, ax1 = plt.subplots(figsize=(12, 8))

# Plot the first line on the first y-axis
ax1.plot(sales_by_date['Transaction_Date'], sales_by_date['TotalAmount'], marker='o', linestyle='-',color='g', label='Total Amount')
ax1.set_xlabel('Transaction Date')
ax1.set_ylabel('Total Amount', color='g')
ax1.tick_params(axis='y', labelcolor='g')
ax1.tick_params(axis='x', rotation=45)
for i, value in enumerate(sales_by_date['TotalAmount']):
    plt.text(sales_by_date['Transaction_Date'][i], value, round(value, 2), ha='center', va='bottom')


# Create the second y-axis and plot the second line
ax2 = ax1.twinx()
ax2.plot(qty_by_date['Transaction_Date'], qty_by_date['Quantity'], marker='*', linestyle='--',color='b', label='Quantity')
ax2.set_ylabel('Quantity', color='b')
ax2.tick_params(axis='y', labelcolor='b')
for i, value in enumerate(qty_by_date['Quantity']):
    plt.text(qty_by_date['Transaction_Date'][i], value, round(value, 2), ha='center', va='bottom')

plt.title('Total Amount & Quantity Over Time')
# Add legends
fig.legend(loc="upper right",bbox_to_anchor=(0.9, 0.88))

# Show the plot
plt.show()
```

#### Observation
##### Higheset Sales by Amount occur on 5-Aug-24 but sold quantity is less, Highest sales occur on 2-Aug-24

### Qty sold by Product Category

```python
category_sales_qty=df.groupby('ProductCategory')['Quantity'].sum()
category_sales_qty
```

```python
# Function to display both percentage and actual amount
def autopct_format(values):
    def my_autopct(pct):
        total = sum(values)
        absolute = int(round(pct*total/100.0))
        return f"{pct:.1f}%\n({absolute})"
    return my_autopct
```

```python
plt.pie(category_sales_qty, labels=category_sales_qty.index, autopct=autopct_format(category_sales_qty.values), startangle=140)
plt.title('Quantity Sold by Product Category')
plt.show()
```

#### Observation
##### Maximum product sold belongs to Toys Category while least products sold in Fashion Category

### Payment Method Usage

```python
payment_counts = df['PaymentMethod'].value_counts()
payment_counts
```

```python
plt.pie(payment_counts, labels=payment_counts.index,autopct=autopct_format(payment_counts.values), startangle=140)
plt.title('Payment Method Distribution')
plt.show()
```

### customers by spending

```python
sales_amount_by_customers=df.groupby('CustomerID')['TotalAmount'].sum().reset_index()
sales_amount_by_customers
```

```python
customer_desc=sales_amount_by_customers.sort_values(by='TotalAmount',ascending=False)
customer_desc
```

```python
plt.figure(figsize=(10, 6))
bars = plt.barh(customer_desc['CustomerID'], customer_desc['TotalAmount'], color='Green')
for bar in bars:
    plt.text(bar.get_width() + 70, bar.get_y() + bar.get_height()/2, round(bar.get_width(), 0),
             ha='center', va='center')
# Invert y-axis to have the highest values on top
plt.gca().invert_yaxis()

plt.title('Customers Spending')
plt.xlabel('Spending')
plt.ylabel('Customer ID')
plt.tight_layout()
plt.show()
```

### Hourly Transaction Volume

```python
df['Transaction_Time'] = pd.to_datetime(df['Transaction_Time'], format='%H:%M:%S').dt.hour
hourly_sales = df.groupby('Transaction_Time')['TotalAmount'].sum().reset_index()
hourly_sales
```

```python
plt.figure(figsize=(10,10))
sns.lineplot(data=hourly_sales, x='Transaction_Time', y='TotalAmount',marker='o')
for i, value in enumerate(hourly_sales['TotalAmount']):
    plt.text(hourly_sales['Transaction_Time'][i], value, round(value, 0), ha='center', va='bottom')
plt.title('Hourly Transaction Volume')
plt.xlabel('Hour of the Day')
plt.ylabel('Total Amount')
plt.show()
```

#### Observation
##### Highest Transaction volume is around 2:00 pm
